```python
import numpy as numpy
import os
import json
from shutil import copyfile

dir_path = '/content/data/NAF_dataset/groups/'


with open('train_valid_test_split.json','r') as json_file: 
    data = json.load(json_file) 

if not os.path.exists('test_set'):
    os.makedirs('test_set')

os.chdir('test_set')
for dir_name in data['test']:
    print(dir_name)

    if not os.path.exists(dir_name):
        os.makedirs(dir_name)

    os.chdir(dir_name)

    for img_name in data['test'][dir_name]:
        print(img_name)
        copyfile(dir_path+dir_name+'/'+img_name,os.getcwd()+'/'+img_name)


    os.chdir('../../')    
```

```python
import os 
import json

folder_name = '/content/data/NAF_dataset/' 
os.chdir(folder_name)
json_file = dict()
json_file = {
    "textBBs": [],
    "fieldBBs": [],
    "pairs": [],
    "samePairs": []
}
for img_name in os.listdir():
    img_name = img_name.split('.')
    json_name = img_name[0]+'.json'
    with open(os.path.join(folder_name,json_name),'w') as new_f:
        json.dump(json_file,new_f)
```
